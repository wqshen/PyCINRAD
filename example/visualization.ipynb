```python
import cinrad
%matplotlib inline
```

**This notebook mainly introduces kwargs and bounded methods of `cinrad.visualize.PPI`**

```python
f = cinrad.io.CinradReader('Z_RADR_I_Z9200_20190421190600_O_DOR_SA_CAP.BIN.BZ2')
```

```python
r = f.get_data(0, 230, 'REF')
```

Use default settings

The `dpi` kwarg is still used to limit the size of this notebook

```python
fig = cinrad.visualize.PPI(r, dpi=75)
```

Annotate city names on the plot

```python
fig = cinrad.visualize.PPI(r, dpi=75, add_city_names=True)
```

Plot range rings

```python
fig = cinrad.visualize.PPI(r, dpi=75)
fig.plot_range_rings([50, 100, 150, 200, 230])
```
